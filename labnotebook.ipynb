```python
from pathlib import Path
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

def size_filter(img: np.ndarray, thresh: int):
    n, labels, stats, _ = cv.connectedComponentsWithStats(img)
    
    mask = np.zeros((*labels.shape, 3), dtype=np.uint8)
    count = 0
    for i in range(n):
        area = stats[i,cv.CC_STAT_AREA]
        if 100000 > area > thresh:
            count += 1
            mask[labels == i,:] = np.random.randint(0, 256, 3, dtype=np.uint8)

    return count, mask

def open(img: np.ndarray, kernel_size: int = 3):
    x = img.copy()
    x = cv.erode(x, cv.getStructuringElement(cv.MORPH_ELLIPSE, [kernel_size]*2))
    x = cv.dilate(x, cv.getStructuringElement(cv.MORPH_ELLIPSE, [kernel_size]*2))
    return x

def dilate_rc(marker: np.ndarray, mask: np.ndarray):
    binary = True if len(np.unique(marker)) == 2 else False

    marker_old = None
    kernel = cv.getStructuringElement(cv.MORPH_RECT, (3, 3))
    while not np.array_equal(marker, marker_old):
        marker_old = np.copy(marker)
        marker = cv.dilate(marker, kernel)
        if binary:
            marker = cv.bitwise_and(marker, mask)
        else:
            marker = np.minimum(marker, mask)

    return marker

def double_threshold(src, upper_narrow, upper_wide):
    marker = cv.inRange(src, 0, upper_narrow)
    mask = cv.inRange(src, 0, upper_wide)
    return dilate_rc(marker, mask)

y_paths = list(Path("./img").glob("*y-neg*")) + list(Path("./img").glob("*y-pos*"))

y = [cv.imread(str(path), cv.IMREAD_COLOR_RGB) for path in y_paths]
masks = []
counts = []
for i in range(len(y)):
    img = cv.cvtColor(y[i], cv.COLOR_RGB2GRAY)
    img[img < 100] = 255 # set black corners white
    clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(19,19)) # adaptive histogram equalization
    img[:] = clahe.apply(img)
    img[:] = double_threshold(img, 100, 150)
    #img[:] = cv.erode(img, cv.getStructuringElement(cv.MORPH_ELLIPSE, [3,3]))
    count, mask = size_filter(img, 300)
    masks.append(mask)
    counts.append(count)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20,10))
ctr = 0
for i in range(2):
    for j in range(2):
        axes[i,j].imshow(y[ctr])
        axes[i,j].set_title(f"count={counts[ctr]}")
        axes[i,j].axis('off')
        axes[i,j+2].imshow(y[ctr+4])
        axes[i,j+2].set_title(f"count={counts[ctr+4]}")
        axes[i,j+2].axis('off')
        ctr += 1
plt.subplots_adjust(wspace=0, hspace=0)
```

We wanted to pipette 200 μl of cell suspension into each Mattek plate and for this we calculated what final volume is needed to achieve same confluency as the original 10 cm plate. Ratio of plate areas was used as the dilution coefficient:
$$\textrm{Resuspension volume} = \frac{\textrm{Area[10 cm plate]}}{\textrm{Area[Mattek plate]}} \times 0.2 \textrm{ml}= \frac{\pi (\textrm{5 cm})^2}{\pi (\textrm{0.7 cm})^2} \times 0.2 \textrm{ml} \approx 10.2 \textrm{ml}$$

In order to achieve 40% confluency on the Mattek plates, we calculated the required final volume of cell suspension in MEGM adjusted for cell loss :
$$\textrm{Final volume} = \frac{\textrm{Resuspension volume}}{\textrm{Target confluency}/\textrm{Starting confluency}}\times \textrm{Total cells left} = \frac{10.2\textrm{ml}}{\textrm{40\%/95\%}}\times 0.6 \times 0.75=10.9 \textrm{ml}$$
