#### 24.11.2025

We prepared the DMEM with penicillin and streptomycin (PS) media (p. 16) by mixing 500 ml DMEM and 5 ml PS which had been warmed to 37°C. From this, three 80 ml aliquots were prepared in glass bottles. All the bottles were refrigerated at 4°C.
</div>
<br>
<div style="border-left: 5px solid #f43f5e; padding-left: 1em;">
<b>Passaging of HEK293 cells</b>

HEK293 cell culture plate (prepared by course personnel) was given to us. We confirmed with bright-field microscope that the cells are adhered to the plate. Old media was aspirated, 1 ml trypsin-EDTA pipetted onto the plate, and incubated for 1 min 37°C. Cell detachment was confirmed with microscope. 5 ml of full DMEM (p. 6; prepared by other students) was added to the plate and mixed by pipetting back and forth. The cell suspension was then centrifuged (300 g, 5 min) in a 15 ml Falcon tube. Supernatant was discarded and the pellet resuspended in 500 µl of full DMEM.

Two cell culture plates were prepared:
<ul>
<li>6 cm plate for transfection: 5 ml full DMEM, 100 µl cell suspension</li>
<li>10 cm plate for maintenance: 10 ml full DMEM, 250 µl cell suspension</li>
</ul>
The plates were stored in the incubator at 37°C ~13:00. Leftover cell suspension was discarded. 
</div>

#### 25.11.2025
<div style="border-left: 5px solid #10b981; padding-left: 1em;">
<b>Plating of hMEC cells</b>

10 ml of DMEM+PS prepared yesterday and 10 ml of mammary epithelial cell growth medium (MEGM; p. 16; prepared by other students) were pipetted into 15 ml Falcon tubes and warmed to 37°C. Vial containing hMEC cells (passage 22) in a ~1 ml volume was fetched from a -69°C freezer and thawed on 37°C water bath. hMEC cells were then mixed with the DMEM+PS and centrifuged (350 g, 5 min). Supernatant was aspirated and the pellet resuspended in 10 ml MEGM. Cell suspension was pipetted onto a 10 cm plate and stored in a 37°C incubator at ~11:00.
</div>
<br>
<div style="border-left: 5px solid #f43f5e; padding-left: 1em;">
<b>Transfection of HEK293 cells</b>

For this work, we used a 6 cm HEK293 cell culture plate prepared by course personnel and not the one made yesterday due to wrong cell line.

1.67 µl of plasmid (pIRESneo-EGFP-alpha Tubulin; 1.2 µg/µl) was mixed with 200 µl jetPRIME buffer and vortexed. 8 µl jetPRIME reagent was pipetted into the mixture, vortexed, and centrifuged briefly. Plasmid mixture was then incubated for 10 min RT. Old media was aspirated from the 6 cm plate and replaced with 5 ml of transfection media (p. 6; prepared by other students). Plasmid mixture was then pipetted onto the plate and shaken lightly. Plate was left in the incubator at 37°C ~12:00.
</div>
