<b>Passaging of HEK293 cells</b>

HEK293 cell culture plate (prepared by course personnel) was given to us. We confirmed with bright-field microscope that the cells are adhered to the plate. Old media was aspirated, 1 ml trypsin-EDTA pipetted onto the plate, and incubated for 1 min 37°C. Cell detachment was confirmed with microscope. 5 ml of full DMEM (p. 6; prepared by other students) was added to the plate and mixed by pipetting back and forth. The cell suspension was then centrifuged (300 g, 5 min) in a 15 ml Falcon tube. Supernatant was discarded and the pellet resuspended in 500 µl of full DMEM.

Two cell culture plates were prepared:
<ul>
<li>6 cm plate for transfection: 5 ml full DMEM, 100 µl cell suspension</li>
<li>10 cm plate for maintenance: 10 ml full DMEM, 250 µl cell suspension</li>
</ul>

<div style="border-left: 5px solid #0ea5e9; padding-left: 1em;">
<b>Isolation and plating of intestinal crypts</b>

150 ml of PBS with 10 mM EDTA was prepared from 0.5 M EDTA by pipetting $\frac{150 \textrm{ml}\times 10 \textrm{mM}}{500 \textrm{mM}} = 3 \textrm{ml}$ into glass bottle and filling to 150 ml with PBS.

Mouse was dissected and intestine harvested according to manual protocol (p. 26).

The intestine was flushed with cold PBS, split apart longitudinally, and rinsed on a petri dish. Then it was cut into pieces on fresh PBS and poured into a 50 ml conical tube. Tissue was washed 3 times in 30 ml (2xPBS, 1xPBS with 10 mM EDTA) by inverting the tube 3 times, letting tissue settle at the bottom, and pouring out supernatant.

To remove villi, the tissue was washed 3 times by adding 30 ml PBS with 10 mM EDTA, incubating on ice (15/15/30 min), vortexing (10/20/60 s) at setting 8.5, letting tissue settle at the bottom, and pouring out supernatant. Supernatant after the second wash was checked for villi with a bright-field microscope (Fig. 3.1). The last supernatant was collected by filtering through a 70 µm cell strainer.

<div style="border-left: 5px solid #f43f5e; padding-left: 1em;">
<b>Selection plating and FACS of transfected HEK293 cells</b>

Full DMEM, PBS, and trypsin-EDTA were warmed to 37°C. Old media was aspirated from the 6 cm transfection plate. Cells were rinsed in 5 ml of PBS and then the PBS was aspirated. 1 ml of trypsin-EDTA was added to the plate and incubated for 1 min in 37°C. Cell detachment was verified with bright-field microscope. Then 5 ml of full DMEM was added to the plate and mixed by pipetting. Cell suspension was centrifuged (300 g, 5 min, RT) and supernatant discarded. Pellet was resuspended in 1000 μl full DMEM.

Next, we prepared a 6-well plate for selection of transfected cells. We prepared 6 ml of full DMEM media with 2 mg/ml G418 from 100 mg/ml stock. $\frac{6 \textrm{ml}\times 2 \textrm{mg/ml}}{100 \textrm{mg/ml}}=120 \textrm{μl}$ of the stock was used. 2 ml of full DMEM was pipetted to wells A1 and B1. 2 ml of full DMEM+G418 was pipetted to A2, B2 and B3. 50 μl of cell suspension was pipetted to well A1 and A2. The plate was left in the incubator.

The remaining cell suspension underwent FACS sorting. $\frac{900 \textrm{μl}\times 1 \textrm{μM}}{1000 \textrm{μM}}=0.9 \textrm{μl}$ of 1 mM SYTOX blue stock was pipetted to the suspension to achieve 1 μM concentration. FITC channel was used to detect EGFP positive or negative cells, and they were sorted into eppendorf tubes containing 500 μl full DMEM. The tubes were centrifuged (300 g, 5 min), supernatant discarded, and pellet resuspended in 300 μl of full DMEM. 150 μl of EGFP+ cells were pipetted into the 6-well plate wells B1 and B2; EGFP- cells into well B3 (Fig. 4.1).
